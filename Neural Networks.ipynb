# Neural Networks

- Functions represented as a graph of computation
- Similar to logistic regression

<img src="files/network-2-3.png" width="400"/>

- Each "neuron" computes a weighted sum plus a constant and applies an "activation" function to comput its output
- This is equivalent to logistic regression when sigmoid is the activation function: $sigmoid(\alpha_0 x_0 + \alpha_1 x_1 + \beta)$

```javascript
var _ = require('lodash');
var loader = require('csv-load-sync');
var d3 = require('d3');
var { JSDOM } = require('jsdom');
var { Layer, Network } = require('synaptic');
var {d3svg, d3bar, d3plot, d3fit, d3decision, argMax} = require('./utilities.js');
```

```javascript
var iris = loader('iris.csv')
speciesColor = {setosa: '#dda0dd', versicolor: '#ffecdb', virginica: '#e08283'};
species = iris.map(o => { return {x: o.Petal_Length, y: o.Petal_Width, color: speciesColor[o.Species]}; });
d3plot(species, 'length', 'width');
```

```javascript
var inputLayer = new Layer(2);
var outputLayer = new Layer(3);

inputLayer.project(outputLayer);

network = new Network({input: inputLayer, hidden: [], output: outputLayer});
```

```javascript
function train(network, data, alpha, N) {
    network.reset();
    for (var i = 0; i < N; i++) {
        for (var sample of data) {
            network.activate(sample.input);
            network.propagate(alpha, sample.output);
        }
    }
}
```

```javascript
var labels = {setosa: [1,0,0], versicolor: [0,1,0], virginica: [0,0,1]};
var trainData = iris.map(o => {
    return {
        input: [parseFloat(o.Petal_Length), parseFloat(o.Petal_Width)],
        output: labels[o.Species]
    };
});
trainData[0]
```

```javascript
train(network, trainData, 0.01, 5000);
```

```javascript
console.log(network.activate(trainData[0].input));
```

```javascript
var labels = ['setosa', 'versicolor', 'virginica'];
var correct = (o) => labels[argMax(network.activate([o.Petal_Length, o.Petal_Width]))] == o.Species;
result = iris.map(o => { return {x: o.Petal_Length, y: o.Petal_Width, color: (correct(o)) ? 'green' : 'red'}; });
d3plot(result, 'length', 'width');
```

```javascript
d3decision((x,y) => argMax(network.activate([x,y])), [1.0,7.5], [0,2.6], ['#dda0dd', '#ffecdb', '#e08283'], species)
```

## Hidden layers

```javascript
var inputLayer2 = new Layer(2);
var hiddenLayer2 = new Layer(5);
var outputLayer2 = new Layer(3);

inputLayer2.project(hiddenLayer2);
hiddenLayer2.project(outputLayer2);

network2 = new Network({input: inputLayer2, hidden: [hiddenLayer2], output: outputLayer2});
```

```javascript
train(network2, trainData, 0.01, 5000);
```

```javascript
var labels = ['setosa', 'versicolor', 'virginica'];
var correct = (o) => labels[argMax(network2.activate([o.Petal_Length, o.Petal_Width]))] == o.Species;
result = iris.map(o => { return {x: o.Petal_Length, y: o.Petal_Width, color: (correct(o)) ? 'green' : 'red'}; });
d3plot(result, 'length', 'width');
```

```javascript
d3decision((x,y) => argMax(network2.activate([x,y])), [1.0,7.5], [0,2.6], ['#dda0dd', '#ffecdb', '#e08283'], species)
```
